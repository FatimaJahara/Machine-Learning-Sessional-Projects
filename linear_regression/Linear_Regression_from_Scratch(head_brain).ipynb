```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (20.0,10.0)

#Reading data
data = pd.read_csv('headbrain.csv')
print(data.shape)
```

```python
data.head()
```

```python
X=data['Head Size(cm^3)'].values
Y=data['Brain Weight(grams)'].values

mean_x=np.mean(X)
mean_y=np.mean(Y)

l=len(X)

num = 0
den = 0
for i in range(l):
    x_diff =(X[i]-mean_x)
    y_diff =(Y[i]-mean_y)
    num+=x_diff*y_diff
    den+=x_diff**2
    
m=num/den
c=mean_y - (m*mean_x)

print("Gradient: ",m) 
print("Intercept: ",c) 
```

```python
sst = 0
ssr = 0
for i in range(l):
    pred_y=m*X[i]+c
    sst+=(Y[i]-mean_y)**2
    ssr+=(Y[i]-pred_y)**2
r2=1-(ssr/sst)

print("R2 Score : ",r2)
```

```python
#max_x = np.max(X)+100
#min_x = np.max(Y)-100

#x=np.linspace(min_x,max_x,1000)
y=m*X+c



plt.plot(X,y,color = '#58b970', label='Regression Line')
plt.scatter(X,Y,color = '#ef5423', label='Scatter Plot')

plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
plt.legend()
plt.show()
```

```python

print("Actual Brain Weight(grams) ----> Predicted Brain Weight(grams)\n")
n = len(Y)
for i in range(n): 
    y_pred=m*X[i]+c
    print("             "+str(Y[i])+"          ---->    "+str(y_pred))
```
